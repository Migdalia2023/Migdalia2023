**Teorema del lmite central**:

______________________

Cuando se selecciona muestras aleatorias de tamao n de una poblacin, la distribucin de muestreo de la media muestral puede aproximarse mediante una distribucin normal a medida que el tamao de muestra se hace grande

Supongamos que tengamos una poblacin con distribucin uniforme entre 0 y 1. Luego generar muestras de 1,10,100 y 1000, y luego ver que se cumple en teorema central de lmite para la estimacin de la media.

Para generar datos con distribucin uniforme se puede usar:

runif(casos,inicio,fin)

Para repetir ms de una vez el experimento:

replicate(veces,runif(casos,inicio,fin))

Vamos por pasos ...

```r
#Generamos 5 nmeros con distribucin uniforme entre 0 y 1
#runif(casos,inicio,fin)
runif(5,0,1)
```

```r
#Podemos repetir el experimento mas de una vez con replicate
#replicate(veces, experimento)
replicate(3,runif(5,0,1))
```

```r
#Nmero de muestras
pruebas<- 10000
#Formamos vectores donde se almacenarn los resultados
muestra1<- vector("numeric",pruebas)
muestra10<- vector("numeric",pruebas)
muestra100<- vector("numeric",pruebas)
muestra1000<-vector("numeric",pruebas)
#Repetimos la muestra en donde se ir generando muestras y se almacenarn las medias
for(i in 1:pruebas){
      m1<-runif(1,0,10)
      m10<-runif(10,0,10)
      m100<-runif(100,0,10)
      m1000<-runif(1000,0,10)
      muestra1[i] <- mean(m1)
      muestra10[i] <- mean(m10)
      muestra100[i] <- mean(m100)
      muestra1000[i]<-mean(m1000)
}
#Luego hacemos las grficas
par(mfrow=c(2,2))
hist(muestra1,main="Muestras de tamao 1")
hist(muestra10,main="Muestras de tamao 10")
hist(muestra100,main="Muestras de tamao 100")
hist(muestra1000,main="Muestras de tamao 1000")
```

**Ejercicio2**:

Generar una muestras de una poblacin normal de media 500 y desviacin 50, los tamaos de la muestra debe ser de 10, 100 y 1000, analizar sus intervalos de confianza.

Para generar la muestra de una poblacin normal se puede usar:

**rnorm(casos,media,varianza)**

```r
#CASO DE LA MUESTRA DE 10 CASOS
#***********************************
n<-10
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Lmite inferior:",Limite_inferior,"\n")
cat("Lmite superior:",Limite_superior,"\n")
```

```r
#CASO DE LA MUESTRA DE 100 CASOS
#***********************************
n<-100
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Lmite inferior:",Limite_inferior,"\n")
cat("Lmite superior:",Limite_superior,"\n")
```

```r
#CASO DE LA MUESTRA DE 1000 CASOS
#***********************************
n<-1000
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Lmite inferior:",Limite_inferior,"\n")
cat("Lmite superior:",Limite_superior,"\n")
```

Generar una muestras de una poblacin normal de media 500 y desviacin 50. Tomar una muestra de 1000, y calcular su intervalo de confianza con:

Se debe instalar el paquete: BSDA


```r
install.packages("BSDA")
```

```r
n<-1000
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Lmite inferior:",Limite_inferior,"\n")
cat("Lmite superior:",Limite_superior,"\n")

library(BSDA)
Resultados<-zsum.test(mean.x=media,sigma.x=desviacion, n.x=1000,conf.level=0.95)
cat("\n","El intervalo de confianza es:")
Resultados$conf.int
```
