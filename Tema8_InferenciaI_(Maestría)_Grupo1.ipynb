**Teorema del límite central**:

______________________

Cuando se selecciona muestras aleatorias de tamaño n de una población, la distribución de muestreo de la media muestral puede aproximarse mediante una distribución normal a medida que el tamaño de muestra se hace grande

**Ejercicio1**:

Supongamos que tengamos una población con distribución uniforme entre 0 y 1. Luego generar muestras de 1,10,100 y 1000, y luego ver que se cumple en teorema central de límite para la estimación de la media.

Para generar datos con distribución uniforme se puede usar:

runif(casos,inicio,fin)

Para repetir más de una vez el experimento:

replicate(veces,runif(casos,inicio,fin))

Vamos por pasos ...

```r
#Generamos 5 números con distribución uniforme entre 0 y 1
#runif(casos,inicio,fin)
runif(5,0,1)
```

```r
#Podemos repetir el experimento mas de una vez con replicate
#replicate(veces, experimento)
replicate(3,runif(5,0,1))
```

```r
#Número de muestras
pruebas<- 10000
#Formamos vectores donde se almacenarán los resultados
muestra1<- vector("numeric",pruebas)
muestra10<- vector("numeric",pruebas)
muestra100<- vector("numeric",pruebas)
muestra1000<-vector("numeric",pruebas)
#Repetimos la muestra en donde se irá generando muestras y se almacenarán las medias
for(i in 1:pruebas){
      m1<-runif(1,0,10)
      m10<-runif(10,0,10)
      m100<-runif(100,0,10)
      m1000<-runif(1000,0,10)
      muestra1[i] <- mean(m1)
      muestra10[i] <- mean(m10)
      muestra100[i] <- mean(m100)
      muestra1000[i]<-mean(m1000)
}
#Luego hacemos las gráficas
par(mfrow=c(2,2))
hist(muestra1,main="Muestras de tamaño 1")
hist(muestra10,main="Muestras de tamaño 10")
hist(muestra100,main="Muestras de tamaño 100")
hist(muestra1000,main="Muestras de tamaño 1000")
```

**Intervalos de confianza para la media**:

Tomando como referencia el teorema de límite central, los intervalos de confianza para la media se calcula:



Formula1.jpg

Error estándar: σ/√n

Margen de error: Z(σ/√n)

Z=1.96 (Nivel de confianza del 95%)

Z=1.65 (Nivel de confianza del 90%)

**Ejercicio2**:

Generar una muestras de una población normal de media 500 y desviación 50, los tamaños de la muestra debe ser de 10, 100 y 1000, analizar sus intervalos de confianza.

Para generar la muestra de una población normal se puede usar:

**rnorm(casos,media,varianza)**

```r
#CASO DE LA MUESTRA DE 10 CASOS
#***********************************
n<-10
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Límite inferior:",Limite_inferior,"\n")
cat("Límite superior:",Limite_superior,"\n")
```

```r
#CASO DE LA MUESTRA DE 100 CASOS
#***********************************
n<-100
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Límite inferior:",Limite_inferior,"\n")
cat("Límite superior:",Limite_superior,"\n")
```

```r
#CASO DE LA MUESTRA DE 1000 CASOS
#***********************************
n<-1000
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Límite inferior:",Limite_inferior,"\n")
cat("Límite superior:",Limite_superior,"\n")
```

Generar una muestras de una población normal de media 500 y desviación 50. Tomar una muestra de 1000, y calcular su intervalo de confianza con:

Se debe instalar el paquete: BSDA


```r
install.packages("BSDA")
```

```r
n<-1000
muestra1<-rnorm(n,500,50)
media<-mean(muestra1)
desviacion<-sd(muestra1)
SE=desviacion/sqrt(n)
valorz=1.96
margen_error=1.96*SE
Limite_inferior<-media-margen_error
Limite_superior<-media+margen_error
cat("margen de error:",margen_error,"\n")
cat("Límite inferior:",Limite_inferior,"\n")
cat("Límite superior:",Limite_superior,"\n")

library(BSDA)
Resultados<-zsum.test(mean.x=media,sigma.x=desviacion, n.x=1000,conf.level=0.95)
cat("\n","El intervalo de confianza es:")
Resultados$conf.int
```

**Ejercicio8**:

Imagínese que se toma una muestra de una población normal de media 300 y deviacion 50. Analizar los resultados de la siguiente prueba de hipótesis:

* H0:μ=300
* H1:μ≠300

Para hacer las pruebas de hipótesis bilateral los comandos a usar son:
t.test(variable_trabajar, mu=valor_probar)

Para las pruebas de hipótesis por colas se usara: alternative="greater" or "less" respecto a la H1

Se usa el método de p valor: es una probabilidad que aporta una medida de la evidencia suministrada por la muestra contra la hipótesis nula. Valores p pequeños indican una evidencia mayor contra Ho.

NO SE RECHAZARÁ HO CUANDO EL P VALOR ES MAYOR A 0.05.

```r
#Tomamos una muestra
muestra=rnorm(1000,300,50)
#Aplicamos el test
t.test(muestra,mu=300)
```

```r
estimacion<-t.test(muestra)
estimacion$conf.int
```
